```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed training data
train_data = pd.read_csv("../data/X_train.csv")
train_labels = pd.read_csv("../data/y_train.csv").squeeze()

# Add risk levels to the data
train_data['Risk Level'] = train_labels.map({0: 'Low Risk', 1: 'High Risk'})  # Adjust for binary labels

# Set up colors and style
plt.style.use('default')
sns.set_style("whitegrid")
colors = ['#2ecc71', '#e74c3c']  # Green = Low Risk, Red = High Risk



# 2. Temperature Distribution by Risk Level
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_data, x='Risk Level', y='temperature', palette=colors)
plt.title('Temperature Distribution by Risk Level')
plt.xlabel('Risk Level')
plt.ylabel('Temperature (°F)')
plt.tight_layout()
plt.show()

# 3. Wind Speed vs Temperature by Risk Level
plt.figure(figsize=(10, 6))
sns.scatterplot(data=train_data, x='temperature', y='wind_speed', hue='Risk Level', palette=colors, alpha=0.6)
plt.title('Wind Speed vs Temperature by Risk Level')
plt.xlabel('Temperature (°F)')
plt.ylabel('Wind Speed (km/h)')
plt.legend(title='Risk Level')
plt.tight_layout()
plt.show()

# 4. Feature Correlation Heatmap
plt.figure(figsize=(8, 6))
features = ["temperature", "wind_speed", "rainfall"]  # Add 'humidity' if available
correlation_matrix = train_data[features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlations')
plt.tight_layout()
plt.show()
```
